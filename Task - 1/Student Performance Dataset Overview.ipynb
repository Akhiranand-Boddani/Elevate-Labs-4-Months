```python
# ==============================
# SECTION 1: ENVIRONMENT SETUP
# ==============================

import pandas as pd
import numpy as np

print("[INFO] Core libraries (pandas, numpy) imported successfully.")
```

```python
# ==============================
# SECTION 2: DATASET LOADING
# ==============================

DATASET_NAME = "Student Performance Dataset"
DATASET_PATH = "../Datasets/StudentsPerformance.csv"

try:
    df = pd.read_csv(DATASET_PATH)
    print(f"[SUCCESS] {DATASET_NAME} loaded successfully from: {DATASET_PATH}")
    print("[VALIDATION] Dataset loaded and ready for inspection.")
except Exception as e:
    print("[ERROR] Failed to load dataset.")
    raise e
```
